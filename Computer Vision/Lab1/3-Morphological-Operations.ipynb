```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img):
    # using cv
    cv.imshow("image", img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    cv.putText(blank_img,text='ABCDE', org=(50, 300),fontFace=cv.FONT_HERSHEY_SIMPLEX, fontScale=5,
                color=(255,255,255) , thickness=25, lineType=cv.LINE_AA)
    return blank_img
```

```python
# def display_img(window_name , img):
#     cv.imshow(window_name, img)
#     cv.waitKey(0)


def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_img()
display_img(img)
```

### Erison

```python
kernel = np.ones((5,5), np.uint8)
erosin1 = cv.erode(img, kernel, iterations=1)
display_img(erosin1)
```

```python
erosin5 = cv.erode(img, kernel, iterations=5)
display_img(erosin5)
```

### Dilation

```python
kernel = np.ones((5,5), np.uint8)
dilation1 = cv.dilate(img, kernel, iterations=5)
display_img(dilation1)
```

### Opening

```python
## Opening = erosin + dilation
img = load_img()
white_noise = np.random.randint(low=0, high=2, size=(600,600))
white_noise
```

```python
white_noise = white_noise*255
display_img(white_noise)
```

```python
noise_img = white_noise + img 
display_img(noise_img)
```

```python
opening = cv.morphologyEx(noise_img, cv.MORPH_OPEN, kernel)
```

```python
display_img(opening)
```

```python
img  = load_img()
black_noise = np.random.randint(low=0, high=2, size=(600,600))
black_noise = black_noise*-255
```

```python
black_noise_img = black_noise + img 
display_img(black_noise_img)
```

```python
black_noise_img[black_noise_img==-255] = 0 
display_img(black_noise_img)
```

```python
closing = cv.morphologyEx(black_noise_img, cv.MORPH_CLOSE,kernel )
display_img(closing)
```
